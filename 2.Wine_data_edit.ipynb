```python
import pandas as pd
import csv
```

```python
df = pd.read_csv('vivno_dataset.csv')
df
```

```python
df['Prices'] = df['Prices'].replace({'\,':'','\$':''}, regex=True).apply(lambda x: float(x) if len(x)>0 else 0)
```

```python
df['year'] = df['Names'].replace({'(?!\d+$).':''},regex=True)
df['year'] = df['year'].apply(lambda x: int(x) if len(x)>0 else 0)
df['Names'] = df['Names'].replace({'\d+$':''}, regex=True)
```

```python
brand_country = pd.DataFrame(df.Countrys.str.split('from',1).tolist(), columns = ['brand','country'], copy=True)
df['brand'] = brand_country['brand']
df['brand'] = df['brand'].apply(str.rstrip)
```

```python
df['location'] = brand_country.country.replace({'[\w\W]+\,.':''}, regex=True)
df = df.dropna()
df['location'] = df['location'].apply(str.lstrip)
```

```python
city = brand_country.country.replace({'\w+$':''}, regex=True)

def func(x):
    try:
        if len(x)>1:
            x.pop(-1)
            return x
        elif len(x)==1:
            x.pop(0)
            return x
    except:
        return 0
def func1(x):
    try:
        if len(x)>1:
            return x[-1]
        elif len(x)==1:
            return x[0]
        else:
            return 0
    except:
        return 0
df['city'] = brand_country.country.str.split(',').apply(func).apply(func1)
df['city'] = df['city'].replace({0:'_Not_given_'})
df['city'] = df['city'].apply(str.lstrip)
```

```python
df['raters_numb'] = df.rates.apply(lambda x: len(x.split()) if "0" not in x else 0)
```

```python
df.drop(df.loc[df['color_wine']=='0'].index, inplace=True)
df.drop(df.loc[df['color_wine']=='Green Wine'].index, inplace=True)
df.drop(df.loc[df['color_wine']=='Collectible'].index, inplace=True)
df.drop(df.loc[df['color_wine']=='Boutique'].index, inplace=True)
df.drop(df.loc[df['color_wine']=='Screw Cap'].index, inplace=True)
df.drop(df.loc[df['color_wine']=='Great Gift'].index, inplace=True)
```

```python
df.to_csv('final_wine_df.csv',index=False)
```
